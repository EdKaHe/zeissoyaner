```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
```

```python
#import a csv with pre nd post coordinates of the markers
df = pd.read_csv("360274_R.csv", sep = ";")
df.sort_values(by = ["y_pre", "x_pre"], ascending = [False, True], inplace = True)
df
```

```python
#get gradient in y-direction between the states
df["dx"]= (df["x_pre"] - df["x_post"])
df["dy"] = (df["y_pre"] - df["y_post"])

#remove rows that have big deviations
df = df[df["dx"].abs() < 1.1*df["dx"].abs().mean()]
df = df[df["dy"].abs() < 1.1*df["dy"].abs().mean()]

df
```

```python
#get the major marks
major = df[df["sz_pre"] == 1]
minor = df[df["sz_pre"] == 0]

#get the shift of the pre and post
dx = (major["x_pre"] - major["x_post"]).values
dy = (major["y_pre"] - major["y_post"]).values

print(dx)
print(dy)
```
